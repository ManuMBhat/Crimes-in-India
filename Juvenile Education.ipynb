```python
#importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys

%matplotlib inline
```

```python
#reading the csv
edu=pd.read_csv("18_01_Juveniles_arrested_Education.csv")
#Categorical data and data is already clean
```

<h1>Obtain data on the educational background of juveniles arrested</h1>

<h3>Comparing the education levels</h3>

```python
#making data more presentable
plt.subplots(figsize=(15,6))
bar_plot_data=[]
bar_plot_data.append(np.sum(edu['Education_Illiterate']))
bar_plot_data.append(np.sum(edu['Education_Upto_primary']))
bar_plot_data.append(np.sum(edu['Education_Above_Primary_but_below_Matric_or_Higher_Secondary']))
bar_plot_data.append(np.sum(edu['Education_Matric_or_Higher_Secondary_&_above']))
plt.bar(['Illiterate','Education Upto Primary','Education upto Secondary','Above Secondary'],bar_plot_data,width=0.35)
plt.title("Comparision of Education levels of Juvenile delinquents from 2001-2010")
plt.xlabel("Education Level")
plt.ylabel("Number of Juveniles")
plt.xticks(rotation=90)
plt.show()
```

We can conclude that: *Most juvenile delinquents drop out of Primary school*

<h3>Finding out proportion of illiterate delinquents</h3>

```python
plt.subplots(figsize=(10,10))
pie_chart_data=[]
pie_chart_data.append(np.sum(edu['Education_Illiterate']))
pie_chart_data.append(np.sum(edu['Education_Total']-edu['Education_Illiterate']))
plt.pie(pie_chart_data,labels=['Illiterate','Educated'],colors=['red','skyblue'],explode=(0.1,0),shadow=True,autopct='%1.1f%%')
plt.title("Educated vs. Illiterate in Juvenile delinquents from 2001-2010")
plt.show()
```

We can conclude that: *<b>25.9%</b> of juvenile delinquents are Illiterate* in 2001 to 2010<br>
<a href="https://knoema.com/atlas/India/topics/Education/Literacy/Youth-literacy-rate">The youth literacy rate of India increased from 76.4 to 86.1 from 2001 to 2011</a>

<h3>Constructing Heatmap to observe the difference between States</h3>

```python
edu_by_state = edu.groupby('Area_Name').sum()
edu_by_state.drop('Year' , axis =1 , inplace = True)
edu_by_state.sort_values(by = 'Education_Total', ascending =False)
plt.subplots(figsize = (10,10))
Heatmap = sns.heatmap(edu_by_state.drop('Education_Total',axis=1), cmap='Reds')
Heatmap.set_xlabel('Education levels')
Heatmap.set_ylabel("State")
Heatmap.set_title('Statewise Education level of juvenile delinquents')
```

As we can observe, *Maharashtra and Madhya Pradesh* have a high number of juvenile delinquents

<h3>Comparing Data of Tamil Nadu and Kerala</h3>


Tamil Nadu and Kerala and Neighbouring states with drastically different literacy rates.<br>
Let us compare their juvenile delinquency rates.


```python
#Obtaining data of Tamil Nadu from dataset
edu_tn=edu[edu['Area_Name']=='Tamil Nadu']
edu_tn
```

```python
plt.subplots(figsize = (15,6))
edu_tn_by_year=edu_tn.groupby('Year').sum()
disp = edu_tn_by_year['Education_Illiterate'].plot(legend=True, label='Illiterate')
disp = edu_tn_by_year['Education_Upto_primary'].plot(legend=True, label='Upto Primary')
disp = edu_tn_by_year['Education_Above_Primary_but_below_Matric_or_Higher_Secondary'].plot(legend=True, label='Upto Secondary')
disp = edu_tn_by_year['Education_Matric_or_Higher_Secondary_&_above'].plot(legend=True, label='Above Secondary')
disp.xaxis.set_ticks(np.arange(2001,2011,1))
disp.set(xlabel = "Year", ylabel = 'Juvenile delinquents in Tamil Nadu', 
       title = 'Juvenile Delinquents throughout the years 2001 to 2010 in Tamil Nadu')
plt.show()
```

```python
#Obtaining data of Kerala from dataset
edu_kr=edu[edu['Area_Name']=='Kerala']
edu_kr
```

```python
plt.subplots(figsize = (15,6))
edu_kr_by_year=edu_kr.groupby('Year').sum()
disp = edu_kr_by_year['Education_Illiterate'].plot(legend=True, label='Illiterate')
disp = edu_kr_by_year['Education_Upto_primary'].plot(legend=True, label='Upto Primary')
disp = edu_kr_by_year['Education_Above_Primary_but_below_Matric_or_Higher_Secondary'].plot(legend=True, label='Upto Secondary')
disp = edu_kr_by_year['Education_Matric_or_Higher_Secondary_&_above'].plot(legend=True, label='Above Secondary')
disp.xaxis.set_ticks(np.arange(2001,2011,1))
disp.set(xlabel = "Year", ylabel = 'Juvenile delinquents in Kerala', 
       title = 'Juvenile Delinquents throughout the years 2001 to 2010 in Kerala')
plt.show()
```

Please make note of the <b>*Y-axis*</b> in the above graphs<br>
We can see a spike at 2003 which we can infer was due to the <a href='https://en.wikipedia.org/wiki/Marad_massacre'>Marad Massacre</a>.<br>
The above data has low number of illiterate delinquents due to Kerala having the highest literacy rate in India. 

```python
plt.subplots(figsize = (15,6))
disp = edu_tn_by_year['Education_Total'].plot(legend=True, label='Tamil Nadu')
disp = edu_kr_by_year['Education_Total'].plot(legend=True, label='Kerala')
disp.xaxis.set_ticks(np.arange(2001,2011,1))
disp.set(xlabel = "Year", ylabel = 'Juvenile delinquents in the 2 states', 
       title = 'Juvenile Delinquents throughout the years 2001 to 2010 in the 2 states')
plt.show()
```

We can observe that there is a decrease in juvenile delinquency in Tamil Nadu in accordance with its *increasing literacy rate*
